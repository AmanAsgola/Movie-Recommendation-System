```python
import numpy as np
```

```python
X=np.array([1,2,4,6,8])
y=np.array([3,4,8,10,15])
X_mean=np.mean(X)
y_mean=np.mean(y)
```

```python
x_diff=X_mean-X
y_diff=y_mean-y
print(x_diff)
print(y_diff)
```

```python
m=round(np.sum(x_diff*y_diff)/np.sum(x_diff**2),2)
```

```python
print(m)
```

```python
c=y_mean-m*X_mean
```

```python
print(c)
```

```python
y_line=m*X+c
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X, y)
plt.plot(X, y_line)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Regression Line")

plt.show()
```
